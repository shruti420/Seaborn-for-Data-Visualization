```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
#univariant set of observations just one variable
sns.distplot(tips["total_bill"],kde=False, bins=40)
#kde=kernel density estimation (a line on graph, by deaflut it is true)
#bins= increase the no of rows by increasing no of bins
```

```python
#joint plot= joins two distplot for bivariant data(two variables ata time )
#kind=modify the type of plot that we're trying...
#hex= it is density based..meaning the more data it is accomodating the more dense it looks
#kind="reg" : regression line inside it
#kind="kde": closer the waves are, more dense the data
sns.jointplot(x="total_bill",y="tip",data=tips, kind="hex")
```

```python
#pair plot- it provides joint plot for every combination of columns presents in dataset
#its a bivariant plot
#hue= differnt color for differnt column
#palette=colorskim
sns.pairplot(tips, hue="sex",palette="coolwarm")
```

```python
#rugplot=just draws a dash mark for every point
#univariant plot
sns.rugplot(tips["total_bill"])
```

```python
#distplot
sns.distplot(tips["total_bill"])
```

```python
#kde plot
sns.kdeplot(tips["total_bill"])
```

# Seaborn : Catagorical Plots


```python
import numpy as np
```

```python
#barplot
#std=standard distribution
sns.barplot(x="sex",y="total_bill", data=tips, estimator=np.std)
```

```python
#countplot
#yaxis value is always gonna be count
sns.countplot(x="sex",data=tips)
```

```python
#boxplot
sns.boxplot(x="day",y="total_bill", data=tips,hue="smoker")
#first box=first quartile, middleline=median value(here bill),top box=third quartile bottom line=min value, top line=max value
#dots above it are called wiskers or outliers
```

```python
#violinplot  bulkier part shows the greater distribution/density
sns.violinplot(x="day",y="total_bill", data=tips, hue="sex",split=True)
```

```python
#stripplot
#jitter= keeps the dots from overlapping with eachother
sns.stripplot(x="day",y="total_bill",data=tips,jitter=True,hue="sex")
```

```python
#swarmplot= combination of violin plots into striplots
#but it does ot scale well

sns.swarmplot(x="day",y="total_bill",data=tips)
```

```python
#combine swarmplot and violin plot
sns.swarmplot(x="day",y="total_bill",data=tips,color="black")
sns.violinplot(x="day",y="total_bill",data=tips)
```

```python
#factorplot
sns.factorplot(x="day",y="total_bill",data=tips, kind="violin")
```

# Matrix Plot

```python
import seaborn as sns
%matplotlib inline
flights = sns.load_dataset("flights")
flights.head()
#following is table is not in form of proper matrix as it cannot achive any kind of tangible information by comparing rows with columns
#so to covert into proper matrix we can use correlation or pivot table
```

```python
fp=flights.pivot_table(index="month",columns="year",values="passengers")
```

```python
#heatmap-the lighter the color is , the more denser the passengers have been
#cmap colorscemes= magma,coolwarm
sns.heatmap(fp, cmap="coolwarm",linewidth=1,linecolor="white")
```

```python
#clustermap- this map follows similar colorsceme together instead of serially arranging data
#notice the rows
sns.clustermap(fp,cmap="magma",standard_scale=1)
```

# Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris= sns.load_dataset('iris')
iris.head()
```

```python
#unique species that iris dataset has
iris["species"].unique()
```

```python
sns.pairplot(iris)
```

```python
a=sns.PairGrid(iris)
#you r gonna be in complete control of the subplots you plot
#everything is manual here
#it is gonna take all the  numerical column and grid them all
a.map(plt.scatter)
```

```python
a.map_diag(sns.displot)#for diagonal plots
a.map_upper(sns.kdeplot) # for upper triangle
a.map_lower(plt.scatter) # for lower triangles
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
b=sns.FacetGrid(data=tips, col="time", row="smoker")
b.map(sns.distplot,"total_bill")
```

```python
b.map(plt.scatter,"total_bill","tip")
```
